## Автокореляция

### Автокорреляция from scratch

Хорошая <a href="https://towardsdatascience.com/understanding-autocorrelation-in-time-series-analysis-322ad52f2199">статья</a> на тему автокореляции и как самому написать код для расчета автокореляции.  

````python
def r_pirson(y: np.array, lag: int):
    y_mean = np.mean(y)
    denominator = np.sum(np.power(y - y_mean, 2))
    numerator = np.sum((y[lag:] - y_mean) * ((y[:-lag] if lag else y) - y_mean))
    return numerator / denominator


def autocorrelation(ts: np.array, n_lags: int = None):
    n_lags = len(ts) if not n_lags else n_lags
    auto_corr_r = list()
    for lag in range(n_lags):
        r = r_pirson(ts, lag)
        step_results = (lag, r)
        auto_corr_r.append(step_results)
    return auto_corr_r


def ac_plot(lags: np.array, r: np.array, title: str, figsize: tuple = (14, 4), ax=None):
    if not ax:
        fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(lags, r)
    ax.axhline(y=0, color='k', linestyle = '-')
    for lag, pirson in list(zip(lags, r)):
        ax.plot([lag, lag], [pirson, 0], color='k', linewidth=0.5, linestyle='--')
    ax.set_title(title, loc='right', fontsize=12)
    ax.grid(axis='both')
    if not ax:
        plt.show()
    return


def calculate_all(df: pd.DataFrame, to_plot: bool = False,
                  figsize: tuple = None, n_lags: int = None):
    results = dict()
    if to_plot:
        fig, ax = plt.subplots(nrows=len(df.columns), ncols=1, figsize=figsize if figsize else (14, len(df.columns)*4))
    for i, col in enumerate(df.columns):
        ts = df[col].values
        curr_autocorr = autocorrelation(ts, n_lags)
        results[col] = curr_autocorr
        lags, r = list(zip(*curr_autocorr))
        if to_plot:
            ac_plot(lags, r, col, ax=ax[i])
    if to_plot:
        fig.tight_layout()
    return results
````

````python
res = calculate_all(train.drop('Period', axis=1), n_lags=None, to_plot=True)
````

## Оценка значимости автокрреляции

### Критерий Стьюдента

Значимость автокорреляции при каком-то фиксированном лаге можно проверить с помощью критерия Стьюдента:
- временной ряд $ y^{T} = y_{1}, ... ,y_{T} $
- $H_{0}$ - автокорреляция равна 0 ($ r_{\tau} = 0 $);
- $H_{1}$ - автокорреляция не равна 0 ($ r_{\tau} \neq 0 $);
- статистика $ T(y^{T}) = \frac{|r_{\tau}| \cdot \sqrt{T - \tau - 2}}{\sqrt{1 - r^{2}_{\tau}}} $
- нулевое распределение $ T(y^{T}) \sim St(T - \tau - 2) $ при $H_{0}$

Временной ряд `Revenue`. Пример:

````python
# лаг автокорреляции
t = 5
# рассматриваемый временной ряд
y_t = train.Sales_quantity.values
# размер выборки
n = len(y_t)
# значение критерия Пирсона для автокорреляции с выбранным лагом
r_t = r_pirson(y_t, t)
# число степеней свободы
k = n - t - 2
# значение статистики критерия Стьюдента
st = np.sqrt((r_t**2 * k) / (1 - r_t**2))
print('Проверяемый лаг автокорреляции:\t\t\t{}'.format(t))
print('Размер выборки n:\t\t\t\t{}'.format(n))
print('Число степеней свободы распределения Стьюдента:\t{}'.format(k))
print('Значение автокорреляции для лага {}:\t\t{:.2f}'.format(t, r_t))
print('Значение статистики критерия Стьюдента:\t\t{:.2f}'.format(st))
````

Если значение статистики больше, чем табличное при выбранном уровне `alpha level`, то мы отвергаем нулевую гипотезу. И вероятность того, что она в итоге окажется верной не превышает `alpha level`.

В данном случае значение статистики меньше табличного: $ t_p = 0.52 < 2.002 = t_t $ для выбранного `alpha level` = 0.05 $ \Rightarrow $ мы принимаем нулевую гипотезу $ \Rightarrow $ автокорреляция не значима.

Сезонность $\Rightarrow$ нестационарность

### p-value

Ошибки в работе с нулевой гипотезой делятся на два типа:
1. **Type I error** - мы отвергли нулевую гипотезу, когда на самом деле она верна  
2. **Type II error** - мы приняли нулевую гипотезу, когда на самом деле она НЕ верна

`p-value` - вероятность совершить ошибку первого рода, т.е. отклонить нулевую гипотезу, когда на самом деле она верна. К примеру, мы приходим домой и видим нашу собаку и разбросаный мусор по всей квартире. На собаке ободок от мусорного ведра. Рассмотрим нулевую гипетезу - сосед залез в нашу квартиру, разбросал мусор и на собаку надел ободок от мусорного ведра. Звучит нелепо и `p-value` для такого будет невероятно маленьким. По сути, `p-value` позволяет оценить, насколько нелепо выглядит наша нулевая гипотеза. `alpha level` - порог "нелепости". Предположим, мы отвергли нулевую гипотезу о том, что во всем виноват сосед. Тогда вероятность того, что она на самом деле ВЕРНА (при том что мы ее отвергли) и есть `p-value`. Обычно `alpha level` принимают равным 5% (0.05), как хороший баланс между двумя ошибками I и II рода. Обычно используют следующие `alpha level`: 1% (0.01), 5% (0.05), 10% (0.1). Если вы получаете $\text{p-value} \leq \text{alpha level}$, то вы отвергаете нулевую гипотезу, т.к. нелепые гипотезы мы отвергаем. <a href="https://www.youtube.com/watch?v=oAfpRO5WZ7o">Видео</a> про `p-value`.

### Критерий KPSS

$ \text{H}_{0} $ - ряд стационарен  
$ \text{H}_{1} $ - ряд НЕ стационарен

Параметр `regression`: нулевая гипотеза для KPSS теста:
- “c” : ряд стационарен вокруг некоторой константы
- “ct” : ряд стационарен вокруг тренда (линейного)

В качестве примера оценим ряд `Revenue`.

````python
kpss_results = sm.tsa.stattools.kpss(train.Revenue.values, regression='c', nlags='auto')
print(f'Statistic value:\t{kpss_results[0]}')
print(f'P-value of the test:\t{kpss_results[1]}')
````

Снизу таблица, в которой индексы - выбранный нами `alpha level`, значения - критический порог статистики. Если полученный `Statistic value` больше, чем критический порог для выбранного `alpha level`, то мы отвергаем нулевую гипотезу. <a href="https://www.youtube.com/watch?v=dDOKxRnCovo">Видео</a> про `alpha level`.

````python
print(pd.Series(data=kpss_results[3]))
````

### Критерий Дики-Фуллера

$ \text{H}_{0} $ - ряд НЕ стационарен  
$ \text{H}_{1} $ - ряд стационарен

Если `p-value` больше выбранного порога "нелепости", то мы не можем отвергнуть $ \text{H}_{0} $. Если близко к порогу - необходимо принимать решение на основании значения статистики. Если значение статистики больше критических значений для выбранного уровня доверительного интервала, то $ \text{H}_{0} $ не может быть отвергнута.

````python
adfuller_results = sm.tsa.stattools.adfuller(train.Revenue.values, autolag=None)

print(f'ADF Statistic:\t\t\t\t\t\t\t{adfuller_results[0]}')
print(f'p-value:\t\t\t\t\t\t\t{adfuller_results[1]}')
print(f'Number of lags, used in regression to determine t-statistic:\t{adfuller_results[2]}')
print(f'Number of observations used in the analysis:\t\t\t{adfuller_results[3]}')
print(f'Critical Values:')
for key, value in adfuller_results[4].items():
    print('\t{}:\t{}'.format(key, value))
````

В нашем случае `p-value` больше выбранного нами порога значимости 0.05 (95% доверительный интевал) $ \Rightarrow \text{H}_{0} $ не может быть отвергнута (ряд не стационарен).

## Приведение к стационарности

### Стабилизация дисперсии ряда (логарифмирование, преобразование Бокса-Кокса)

У нас есть некий временной ряд, дисперсия которого изменяется с течением времени $ \Rightarrow $ ряд не стационарный.

````python
x = np.linspace(0, 1000)
y = np.sin(2 * np.pi * x) * (x * 0.05 + 8) + (x * 0.1 + 50)
plt.plot(y)
plt.show()
````

Чтобы стабилизировать дисперсию ряда, можно его прологарифмировать. Получим:

````python
plt.plot(np.log(y))
plt.show()
````

Стало лучше, но не идеально.

Преобразования Бокса-Кокса - общий случай логарифмирования:

$$
y'_{t} = \begin{equation*}
 \begin{cases}
   \ln{y_{t}}, \lambda = 0
   \\
   (y^\lambda - 1)\text{ / }\lambda, \lambda \ne 0
 \end{cases}
\end{equation*}
$$

````python
yt, _ = scs.boxcox(y)
plt.plot(yt)
plt.show()
````

Теперь сравним исходный ряд, рад полученный с помощью логарифмирования и ряд полученный с помощью трансформаций бокса-кокса

````python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16, 4))

prob_1 = scs.probplot(y, dist='norm', plot=ax[0], fit=True)
prob_2 = scs.probplot(np.log(y), dist='norm', plot=ax[1], fit=True)
prob_3 = scs.probplot(yt, dist='norm', plot=ax[2], fit=True)

ax[0].title.set_text('y')
ax[1].title.set_text('log(y)')
ax[2].title.set_text('Box-Cox transformation')
fig.subplots_adjust(wspace=0.4)
fig.show()
````

Формулы обратного преобразования:


$$
\hat{y}_{t} = \begin{equation*}
 \begin{cases}
   \exp({\hat{y}_{t}^{'}}), \lambda = 0
   \\
   (\lambda \hat{y}_{t}^{'} + 1)^{\frac{1}{\lambda}}, \lambda \ne 0
 \end{cases}
\end{equation*}
$$



Особенности:
- не может применяться к рядам, у которых есть нулевые или отрицательные значения. Но можно просто ряд сдвинуть на константу, чтобы значения стали положительными. Потом соответственно эту константу вычесть из прогноза.
- $\lambda$ не особо сильно влияет на преобразование, поэтому можно не стараться ее подбирать. $\lambda$ сильно влияет на предсказательный интервал.

### Q-Q plot

<a href="youtube.com/watch?v=okjYjClSjOg">Видео</a>, где хорошо объяснены Q-Q графики.

Пример:

Сгенерируем выборку из равномерного распределения.

````python
values = np.random.uniform(size=200)
fig, ax = plt.subplots(figsize=(16, 4))
ax.hist(values, bins=100)
plt.show()
````

````python
sample_quantiles_values = np.sort(values)
normal_distribution_quantiles_values = stats.norm.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1,
                                                                  endpoint=False)[1:])
uniform_distribution_quantiles_values = stats.uniform.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1, 
                                                                      endpoint=False)[1:])
chisquare_distribution_quantiles_values = stats.chi2.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1, 
                                                                     endpoint=False)[1:], df=1)
````

Построим Q-Q plots

````python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
ax[0].plot(normal_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='normal')
ax[1].plot(uniform_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='uniform')
ax[2].plot(chisquare_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='chi square')
ax[0].title.set_text('normal')
ax[1].title.set_text('uniform')
ax[2].title.set_text('chi square with df=1')
plt.show()
````

Видим, что для равномерного распределения график приближается в прямой линии, в отличии от остальных. Это значит, что вероятней всего данные пришли из равномерного распределения. Можно также построить линейные регрессии для разультатов каждого из двух распределений и оценить MSE.

Теперь попробуем сделать то же самое, но для данных из нормального распределения.

Сгенерируем данные из нормального распределения.

````python
values = np.random.normal(size=200)
fig, ax = plt.subplots(figsize=(16, 4))
ax.hist(values, bins=100)
plt.show()
````

````python
sample_quantiles_values = np.sort(values)
normal_distribution_quantiles_values = scs.norm.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1,
                                                    endpoint=False)[1:])
uniform_distribution_quantiles_values = scs.uniform.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1, 
                                                        endpoint=False)[1:])
chisquare_distribution_quantiles_values = scs.chi2.ppf(np.linspace(0, 1, len(sample_quantiles_values) + 1, 
                                                       endpoint=False)[1:], df=1)
````

Построим Q-Q plots

````python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
ax[0].plot(normal_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='normal')
ax[1].plot(uniform_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='uniform')
ax[2].plot(chisquare_distribution_quantiles_values, sample_quantiles_values, marker='o', ls='', label='chi square')
ax[0].title.set_text('normal')
ax[1].title.set_text('uniform')
ax[2].title.set_text('chi square with df=1')
plt.show()
````

Видим, что ситуация аналогичная.

В разделе ```3.1.4``` изображены "почти" Q-Q графики. Разница лишь в том, что по оси ординат также расположены значения нашей выборки (значения квантилей), а по оси абсцисс - значения, которые в данном случае вычислены по формуле:

$ \text{quantiles} = \text{dist}.\text{ppf}(\text{val}) $, где

$$
\text{val} = \begin{equation*}
 \begin{cases}
   0.5^{\frac{1}{n}} \text{, для } i = n
   \\
   \frac{i - 0.3175}{n + 0.365} \text{, для } i = 2, ..., n - 1
   \\
   1 - 0.5^{\frac{1}{n}} \text{, для } i = 1
 \end{cases}
\end{equation*}
$$

Из <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.probplot.html">документации scipy</a>:

```
Probplot generates a probability plot, which should not be confused with a Q-Q or a P-P plot. Statsmodels has more extensive functionality of this type, see statsmodels.api.ProbPlot.
```

Т.е. разница заключается в методе расчета квантилей распределения, принадлежность к которому мы хотим проверить наших данных.

К примеру:

````python
def calc_val(i: int, n: int):
    if i == n:
        val = 0.5**(1 / n)
    elif i == 1:
        val = 1 - 0.5**(1 / n)
    else:
        val = (i - 0.3175) / (n + 0.365)
    return val


n = len(y)
for i in range(n):
    val = calc_val(i + 1, n)
    assert scs.norm.ppf(val) == prob_1[0][0][i]
print('Done')
````

### Дифференцирование

Идея заключается в том, чтобы вычесть из значения ряда предыдущее значение и таким образом привести его к стационарному виду (но при этом ряд уменьшается на одно значение). Этим преобразованием мы избавляется от тренда.

Если одного дифференцирования не достаточно, то можно его повторить n-раз.

Пример:

**Критерий KPSS**

````python
kpss_statistic_ini, kpss_pvalue_ini, _, kpss_table = sm.tsa.stattools.kpss(train.Revenue.values,
                                                                           regression='c', nlags='auto')
kpss_statistic_diff, kpss_pvalue_diff, _, _ = sm.tsa.stattools.kpss(train.Revenue.diff().dropna().values,
                                                                    regression='c', nlags='auto')
kpss_table = pd.DataFrame({key: [val] for key, val in kpss_table.items()}, index=['critical_value'])
print('Таблица для оценки стационарности ряда по значению статистики критерия KPSS')
kpss_table
````

````python
print('Значение статистики KPSS исходного ряда:\t\t{:.2f}'.format(kpss_statistic_ini))
print('Значение статистики KPSS продифференцированного ряда:\t{:.2f}'.format(kpss_statistic_diff))
````

Выберем значение `alpha level` == 0.05. Мы видим, что значение статистики KPSS для исходного ряда больше чем значение из таблицы для выбранного значения `alpha level`: $1.22 > 0.463$ $\Rightarrow$ гипотезу $H_{0}$ отвергаем $\Rightarrow$ ряд не стационарный.

Для продифференцированного ряда значение критерия KPSS меньше, чем значение из таблицы для выбранного значения `alpha level`: $0.09 < 0.463$ $\Rightarrow$ гипотезу $H_{0}$ принимаем $\Rightarrow$ ряд стационарный.

**Критерий Дики-Фуллера**

````python
adfuller_statistic_ini, adfuller_pvalue_ini, _, _, adfuller_table_ini = sm.tsa.stattools.adfuller(train.Revenue.values, autolag=None)
adfuller_statistic_diff, adfuller_pvalue_diff, _, _, adfuller_table_diff = sm.tsa.stattools.adfuller(train.Revenue.diff().dropna().values,
                                                                                                     autolag=None)
adfuller_table_ini = pd.DataFrame({key: [val] for key, val in adfuller_table_ini.items()}, index=['critical_value'])
adfuller_table_diff = pd.DataFrame({key: [val] for key, val in adfuller_table_diff.items()}, index=['critical_value'])
````

````python
print('Значение статистики Дики-Фуллера исходного ряда:\t{:.2f}'.format(adfuller_statistic_ini))
print('Значение p-value исходного ряда:\t\t\t{:.2f}'.format(adfuller_pvalue_ini))
print('Таблица для оценки стационарности исходного ряда по значению статистики критерия Дики-Фуллера')
adfuller_table_ini
````

Выберем значение `alpha level` == 0.05. Мы видим, что значение `p-value` выше, чем выбранный `alpha level`: $0.93 > 0.05$ $\Rightarrow$ гипотезу $H_{0}$ отвергнуть мы не можем $\Rightarrow$ ряд не стационарный. Также значение статистики Дики-Фуллера больше значение статистики из таблицы для выбранного `alpha level`: $-0.27 > -2.92$, что также подтверждает то, что ряд не стационарен (гипотезу $H_{0}$ не отвергаем). 

````python
print('Значение статистики Дики-Фуллера продифференцированного ряда:\t{:.2f}'.format(adfuller_statistic_diff))
print('Значение p-value продифференцированного ряда:\t\t\t{:.2f}'.format(adfuller_pvalue_diff))
print('Таблица для оценки стационарности продифференцированного ряда по значению статистики критерия Дики-Фуллера')
adfuller_table_diff
````

Для продифференцированного ряда значение `p-value` меньше `alpha-level`, также как и значение статистики меньше значения статистики для выбранного `alpha-level` $\Rightarrow$ гипотезу $H_{0}$ отвергаем $\Rightarrow$ ряд стационарен

Проилюстрируем продифференцированный ряд на графике:

````python
ts_diff = train.Revenue.diff().dropna().values
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 6))
ax[0, 0].plot(train.Revenue.values)
ax[0, 0].title.set_text('Исходный ряд')
smt.graphics.plot_acf(train.Revenue.values, ax=ax[1, 0])
ax[0, 1].plot(ts_diff)
ax[0, 1].title.set_text('Продифференцированный ряд')
smt.graphics.plot_acf(ts_diff, ax=ax[1, 1])
fig.subplots_adjust(hspace=0.4)
plt.show()
````

# Прогнозирование

<a href="https://towardsdatascience.com/time-series-models-d9266f8ac7b0">Статья</a> про модели ARMA, ARIMA, SARIMA.

## SARIMA

Состоит из двух компонент.

**AR** - авторегрессионная модель, в которой текущее значение $y_{t}$ определяется линейной комбинацией $p$ предыдущих моментов времени (авторегрессия порядка $p$). Модель можно выразить уравнением:

$$ y_{t} = \alpha + \varepsilon_{t} + \phi_1 \cdot y_{t-1} + \ldots + \phi_p \cdot y_{t-p} $$

где 
* $y_{t}$ - текущее значение стационарного ряда
* $y_{t-1}, \ldots, y_{t-p}$ - $p$ предыдущих значений
* $ \phi_1, \ldots \phi_p $ - весовые коэфициенты  модели
* $ \alpha $ - bias
* $ \varepsilon_{t} $ - значение шума в текущий момент времени

**MA (moving average)** - модель скользящего среднего. Это линейная комбинация значений шума в предыдущие $q$ моменты времени. Т.е. по факту, это авторегрессия на шум.

$$ y_{t} = \alpha + \varepsilon_{t} + \theta_1 \cdot \varepsilon_{t-1} + \ldots + \theta_q \cdot \varepsilon_{t-q} $$

где 
* $y_{t}$ - текущее значение стационарного ряда
* $\varepsilon_{t-1}, \ldots, \varepsilon_{t-q}$ - $q$ предыдущих значений шума
* $ \theta_1, \ldots \theta_q $ - весовые коэфициенты  модели
* $ \alpha $ - bias
* $ \varepsilon_{t} $ - значение шума в текущий момент времени

**ARMA** модель выглядит следующим образом:

$$ ARMA(p, q) = \alpha + \varepsilon_{t} + \phi_1 \cdot y_{t-1} + \ldots + \phi_p \cdot y_{t-p} + \theta_1 \cdot \varepsilon_{t-1} + \ldots + \theta_q \cdot \varepsilon_{t-q} $$

**Теорема Вольда**: любой стационарный ряд может быть описан моделью ARMA(p, q) с любой точностью.

### Приведение к стационарному ряду

````python
def tsplot(x: pd.Series, y: pd.Series, title: str, lags: int = None,
           figsize: tuple = (12, 7), style: str = 'bmh', alpha=0.05):
    with plt.style.context(style):
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))

        ts_ax.plot(x, y)
        ts_ax.set_title(title)
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=alpha)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=alpha)

        plt.tight_layout()
        pvalue = sm.tsa.stattools.adfuller(y)[1]
        print('p-value критерия Дики-Фуллера: {:.3f}'.format(pvalue))
    return
````

### Подбор оптимальной модели

````python
ps, ds, qs, Ps, Ds, Qs = range(0, 6), range(0, 2), range(0, 6), range(0, 2), range(0, 2), range(0, 2)
params = product(ps, ds, qs, Ps, Ds, Qs)
params_list = list(params)
print('Количество моделей: {}'.format(len(params_list)))
````

````python
results = list()
best_aic = float('inf')

for param in tqdm(params_list):
    p, d, q, P, D, Q = param
    try:
        model = sm.tsa.statespace.SARIMAX(revenue_df['Revenue_box'], order=(p, d, q),
                                          seasonal_order=(P, D, Q, 12)).fit(disp=-1)
    except:
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append((param, aic))

warnings.filterwarnings('default')
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
````

````python
model = sm.tsa.statespace.SARIMAX(revenue_df['Revenue_box'], order=(2, 1, 0),
                                  seasonal_order=(0, 1, 1, 12)).fit(disp=-1)
print(model.summary())
````

````python
tsplot(revenue_df['Period'], model.resid, title='Остатки', lags=30)
````

Как мы видим, остатки распределены нормально и не имеют значимых автокореляционных лагов.

### Forecasting

````python
fitted_values = inv_boxcox(model.fittedvalues, lmbda)
revenue_df['forecast'] = fitted_values
forecast_period = 2 * 12

forecast_df = pd.DataFrame({'Period': pd.date_range('2020-05-01', periods=forecast_period, freq="1MS")})
forecast = inv_boxcox(model.predict(start=revenue_df.shape[0], end=revenue_df.shape[0] - 1 + forecast_period), lmbda)
forecast_df['forecast'] = forecast.values
df_merged = pd.concat([revenue_df, forecast_df], axis=0, ignore_index=True)
````
