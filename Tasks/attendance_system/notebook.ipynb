```python
from db import read_embeddings
import face_recognition
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
known_face_metadata, known_face_encodings = read_embeddings()
```

```python
frame = cv.imread("./data/elon.jpg")
scale = 0.25
frame_s = cv.resize(frame, (0, 0), None, scale, scale)
frame_s = cv.cvtColor(frame_s, cv.COLOR_BGR2RGB)

face_curr_locations = face_recognition.face_locations(frame_s)
face_encodings_to_check = face_recognition.face_encodings(frame_s, known_face_locations=face_curr_locations)
print("ОБнаружено лиц: %s" % len(face_encodings_to_check))
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.imshow(frame)
for bottom, right, top, left in face_curr_locations:
    x1, y1 = left, bottom
    w, h = right - left, top - bottom
    rect = Rectangle(xy=(x1 / scale, y1 / scale), width=w / scale, height=h / scale, facecolor="none", edgecolor="r", linewidth=1)
    ax.add_patch(rect)
```
