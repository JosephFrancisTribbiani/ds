```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
```

### Загружаем изображение

```python
loc = "./example.jpg"
img = cv.imread(loc)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(img)
ax.axis("off");
```

### Шаг 1. Noise reduction

Делаем с помощью Gaussian blur $5 \times 5$ и $\sigma = 1.4$.

```python
sigma = 1.4
img_fltrd = cv.GaussianBlur(img, (5, 5), 0)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(14, 6))
ax[0].imshow(img)
ax[0].axis("off")

ax[1].imshow(img_fltrd)
ax[1].axis("off");
```
